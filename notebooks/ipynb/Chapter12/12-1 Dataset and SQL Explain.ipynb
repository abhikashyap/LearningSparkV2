```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("LearningSparkV2") \
    .master("local[*]") \
    .getOrCreate()

spark.sparkContext.setLogLevel("WARN")
print(f"Spark {spark.version} ready")
```

### Use `ds.explain("format")`

Format include `simple`, `extended`, `codegen`, `cost`, `formatted`

```python
strings = spark.read.text("/Users/abhikashyap10/spark/LearningSparkV2/databricks-datasets/learning-spark-v2/SPARK_README.md")
filtered = strings.filter(strings.value.contains("Spark"))
filtered.count()
```

```python
filtered.explain(mode="simple")
```

```python
filtered.explain(mode="extended")
```

```python
filtered.explain(mode="cost")
```

```python
filtered.explain(mode="codegen")
```

```python
filtered.explain(mode="formatted")
```

### Use SQL EXPLAIN format

```python
strings.createOrReplaceTempView("tmp_spark_readme")
```

```python
# SQL cell
EXPLAIN FORMATTED SELECT * FROM tmp_spark_readme WHERE value % 'Spark'
```
